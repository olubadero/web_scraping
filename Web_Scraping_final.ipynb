```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import urllib.request
import re
```

```python
## get url
res = requests.get("https://www.jumia.com.ng/category-fashion-by-jumia/?q=princess+dress&size=M&shipped_from=country_local--jumia_global&page=2#catalog-listing")
```

```python
## convert to a beautiful soup object
soup = bs(res.content, "html.parser")
```

```python
jumia = soup.find_all('div', {'class':'info'})
Final = []
print(jumia)
```

```python
for i in jumia:
    name = i.find('h3', {'class':'name'}).get_text()
    shipping = i.find('div', {'class':'bdg _glb _xs'}).get_text()
    price = i.find('div', {'class':'prc'}).get_text()
    
    column = [name, shipping, price]
    Final.append(column)
    
    print(Final)
    
```

```python
jumia_data = pd.DataFrame(Final, columns= ['Name', 'Shipping', 'Price'])
jumia_data
```
